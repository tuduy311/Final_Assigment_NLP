```python
# chọn run -> run current and after
import os, tempfile, time, json, traceback
from pathlib import Path
from threading import Thread
from datetime import datetime, timezone
os.environ["PYTORCH_ALLOC_CONF"] = "expandable_segments:True"
import torch
from flask import Flask, request, jsonify
from faster_whisper import WhisperModel as _FasterWhisperModel

# ── Config ──────────────────────────────────────────

MODEL_PATHS = {
    "whisper-base":"/kaggle/input/models/letrungtin2k5/whisper-base/transformers/default/1/whisper-base",
    "whisper-large-v3":"/kaggle/input/models/letrungtin2k5/systrans-whisper-large-v3/other/default/1/large-v3"
}
MODEL_KEY="whisper-large-v3"
DIARIZATION_MODEL = "/kaggle/input/models/letrungtin2k5/speaker-diarization-community-1/pytorch/default/2/speaker-diarization-community-1"
DUMMY_AUDIO = "/kaggle/input/datasets/letrungtin2k5/toeic-speaking-part3-audio/Test 07 (32-34) (1).mp3"

# Ngưỡng alert
CONFIDENCE_THRESHOLD  = -0.8   # avg_logprob dưới này = kém
NO_SPEECH_THRESHOLD   = 0.6    # trên này = nhiều noise
LATENCY_THRESHOLD_MS  = 36_000 # 36s

STT_DEVICE         = torch.device("cuda:0")
DIARIZATION_DEVICE = torch.device("cuda:1")
LLM_GPU_LAYERS     = -1  # llama-cpp tự map lên cuda:0

class STTModel:
    def transcribe(self, audio_path: str) -> dict:
        raise NotImplementedError
    def warmup(self, audio_path: str):
        self.transcribe(audio_path)

class LLMModel:
    def generate(self, messages, temperature=0.1, max_tokens=1024) -> str:
        raise NotImplementedError
    def warmup(self):
        self.generate("Hello")

class DiarizationModel:
    def diarize(self, audio_path: str) -> list:
        raise NotImplementedError

    def warmup(self, audio_path: str):
        self.diarize(audio_path)

```

```python
class FasterWhisperSTT(STTModel):
    def __init__(self, model_path: str):
        self.model = _FasterWhisperModel(
            model_path,
            device="cuda",
            device_index=0,
            compute_type="float16"
        )

    def transcribe(self, audio_path: str) -> dict:
        segments, info = self.model.transcribe(
            audio_path,
            vad_filter=True,
            temperature=0,
            beam_size=5,
            condition_on_previous_text=False,
            vad_parameters={"min_silence_duration_ms": 500},
            word_timestamps=False
        )
        
        segments = list(segments)

        if not segments:
            return {
                "text": "",
                "language": info.language,
                "language_probability": info.language_probability,
                "confidence": -1.0,
                "no_speech_prob": 1.0,
                "num_segments": 0,
                "duration_sec": 0.0,
            }

        full_segments = []
        result_text = []
        for seg in segments:
            full_segments.append({
                    "start": float(seg.start),
                    "end": float(seg.end),
                    "text": seg.text.strip()
            })
            result_text.append(seg.text.strip())
        text = " ".join(result_text)
        
    
        confidence = sum(seg.avg_logprob for seg in segments) / len(segments)
        no_speech = sum(seg.no_speech_prob for seg in segments) / len(segments)
    
        duration = float(segments[-1].end - segments[0].start)

        return {
            "text": text,
            "segments": full_segments,
            "language": info.language,
            "language_probability": info.language_probability,
            "confidence": confidence,       # avg_logprob, càng gần 0 càng tốt
            "no_speech_prob": no_speech,    # càng thấp càng tốt
            "num_segments": len(segments),
            "duration_sec": duration,
        }


from llama_cpp import Llama

class QwenGGUF(LLMModel):
    def __init__(self, model_path: str):
        self.model = Llama(
            model_path,
            n_ctx=16384,
            n_gpu_layers=-1,
            main_gpu=0,               # ← llama-cpp dùng cuda:0
            verbose=False
        )

    def generate(self, messages, temperature=0.1, max_tokens=1024):
        output = self.model.create_chat_completion(
            messages=messages,
            temperature=temperature,
            max_tokens=max_tokens
        )

        return output["choices"][0]["message"]["content"]

from pyannote.audio import Pipeline
import av
import numpy as np
import torchaudio.functional as Func
import io

class PyannoteSpeakerDiarization(DiarizationModel):
    TARGET_SR    = 16000
    SHORT_SEG_TH = 1.0

    def __init__(self, model_name: str):
        self.device = DIARIZATION_DEVICE
        with torch.cuda.device(self.device):        # ← context manager
            self.pipeline = Pipeline.from_pretrained(model_name)
            self.pipeline.to(self.device)


    # ── Public API ────────────────────────────────────────
    def diarize(self, audio_input: "str | bytes") -> dict:
        """
        Nhận path (str) hoặc raw bytes (mp3/wav/flac/...).
        Luôn convert sang 16kHz mono tensor trước khi inference.
        """
        audio_dict = self._load_audio(audio_input)
        diarization = self.pipeline(audio_dict)
        segments = []
        # community-1 / precision-2 (mới) → output.speaker_diarization
        if hasattr(diarization, "speaker_diarization"):
            for turn, speaker in diarization.speaker_diarization:
                segments.append({"speaker": speaker, "start":  float(turn.start),"end":  float(turn.end),})
        # pyannote 3.1 local (cũ) → itertracks
        elif hasattr(diarization, "itertracks"):
            for turn, _, speaker in diarization.itertracks(yield_label=True):
                segments.append({"speaker": speaker, "start":   float(turn.start), "end":  float(turn.end),})
        else:
            raise RuntimeError(f"Unknown diarization output type: {type(diarization)}")
        return {"segments": segments}

    # ── Private: audio loading & preprocessing ────────────
    def _load_audio(self, audio_input: "str | bytes") -> dict:
        """
        Convert bất kỳ format nào → 16kHz mono torch.Tensor.
        Dùng PyAV (in-process ffmpeg) — không spawn subprocess.
        """
        raw = audio_input if isinstance(audio_input, bytes) else open(audio_input, "rb").read()
        buf = io.BytesIO(raw)

        # Decode bằng PyAV
        waveform, sr = self._av_decode(buf)

        # Resample nếu cần (numpy → torch → resample → giữ torch)
        tensor = torch.from_numpy(waveform)          # (channels, samples)
        if sr != self.TARGET_SR:
            tensor = Func.resample(tensor, sr, self.TARGET_SR)

        # Mono
        if tensor.shape[0] > 1:
            tensor = tensor.mean(dim=0, keepdim=True)

        return {
            "waveform":    tensor.to(self.device),  # ← pin về cuda:1
            "sample_rate": self.TARGET_SR,
        }

    @staticmethod
    def _av_decode(buf: io.BytesIO) -> "tuple[np.ndarray, int]":
        """Decode audio bytes → float32 numpy array dùng PyAV (in-process)."""
        frames, sr = [], None
        with av.open(buf) as container:
            stream = container.streams.audio[0]
            sr     = stream.codec_context.sample_rate
            for frame in container.decode(stream):
                arr = frame.to_ndarray()          # (channels, samples) int16/float
                if arr.dtype != np.float32:
                    arr = arr.astype(np.float32) / np.iinfo(np.int16).max
                frames.append(arr)

        waveform = np.concatenate(frames, axis=-1)  # (channels, total_samples)
        return waveform, sr

```

```python
SUMMARY_SYSTEM = """
You are an expert meeting summarization assistant.
Analyze the transcript and return markdown only.

Use exactly this structure:

# Meeting Summary

## Meeting Information
Extract if available:
- title
- date
- participants
- other information

If missing, omit the field.

---

## Summary
Provide a concise overall summary of the meeting.

---

## Key Decisions
List important decisions made during the meeting.

---

## Discussion Highlights
List major discussion topics and important points raised.

Formatting rules:
- Output markdown only
- No JSON
- No explanations outside markdown
- Keep concise and structured
"""


TASK_SYSTEM = """
You are an action item extraction engine.
Analyze the meeting transcript and extract all actionable follow-up tasks.
Return ONLY valid JSON. No markdown fences, no text before or after.
Schema:
{
  "action_items": [
    {
      "title": "string or null",
      "assignees": ["string"] or [],
      "description": "string",
      "deadline": {
        "raw_phrase": "exact words from transcript or null",
        "reasoning": "brief explanation of what you found and why you classified it this way",
        "anchor": {
          "type": "meeting_date | absolute_in_transcript | event_in_transcript | unknown",
          "absolute_value": "e.g. 'March 10th' if type is absolute_in_transcript, else null"
        },
        "offset_from_anchor": "e.g. '+7 days', 'next Friday', '-1 day then +7 days', or null",
        "resolved": null,
        "confidence": "high | low | unresolvable"
      } or null,
      "note": "string or null",
      "reference_segments": [{"start": number, "end": number}] or null
    }
  ]
}
Rules:
- Return valid JSON only.
- No markdown, no explanation text outside JSON.
- Unknown values must be null.
- Only include clear, actionable commitments.
- 'assignees': SHORT proper names only of the people DOING the task. 
  CRITICAL: Do not assign tasks to the beneficiary or subject of the discussion. For example, if the transcript says "I will talk to John", the task is NOT assigned to John.
  NO pronouns (I, He, We, They). If the person doing the task is only referred to by a pronoun or is completely unnamed in the text, you MUST return [].
- 'deadline': MUST be an object following the schema above, or null if no deadline is mentioned.
  - Step 1: Write 'reasoning' first based on what the transcript actually says.
  - Step 2: Classify 'anchor.type': use 'meeting_date' for relative expressions, 'absolute_in_transcript' for explicit dates, 'event_in_transcript' for event-based anchors, 'unknown' if unclear.
  - Step 3: Fill 'anchor.absolute_value' ONLY if type is 'absolute_in_transcript'.
  - Step 4: Set 'offset_from_anchor' to the relative offset expression.
  - Step 5: Always leave 'resolved' as null — the backend will compute this.
  - Step 6: Set 'confidence': 'high' if anchor is clear, 'low' if ambiguous, 'unresolvable' if impossible.
  - CRITICAL: NEVER invent a YYYY-MM-DD date. 'resolved' is ALWAYS null.
- 'reference_segments': MAX 3 segments using numeric timestamps from transcript.
- Deduplication: If a task is rescheduled/updated mid-conversation, output ONE task with the FINAL state and all relevant timestamps in 'reference_segments'.
"""

CHECK_CONFLICTS_SYSTEM = """
You are a calendar conflict detector.
You will be given a list of candidate pairs representing a new action item (task) and an existing calendar event.
Compare each pair and determine if they refer to the same logical task.

Rules:
- A 2-day or less date difference with same topic = DUPLICATE
- Same topic, different dates beyond 2 days = RELATED
- Omit DIFFERENT pairs entirely

Return ONLY valid JSON in this format — no markdown, no explanation:
{
  "conflicts": [
    {
      "task_id": <int>,
      "event_id": "<string>",
      "verdict": "DUPLICATE" | "RELATED",
      "reason": "<one concise sentence>",
      "suggested_action": "skip" | "ask_reschedule"
    }
  ]
}
"""

```

```python
import uvicorn
from fastapi import FastAPI, File, UploadFile, HTTPException
from fastapi.responses import JSONResponse
from fastapi.responses import StreamingResponse
from pydantic import BaseModel
from concurrent.futures import ThreadPoolExecutor
import asyncio
from huggingface_hub import hf_hub_download

# ── khởi tạo models (giữ nguyên) ─────────────────────────────────────────────
stt_model = FasterWhisperSTT(MODEL_PATHS[MODEL_KEY])
diarization_model = PyannoteSpeakerDiarization(DIARIZATION_MODEL)

repo  = "Qwen/Qwen2.5-14B-Instruct-GGUF"
file1 = hf_hub_download(repo, "qwen2.5-14b-instruct-q5_k_m-00001-of-00003.gguf")
hf_hub_download(repo, "qwen2.5-14b-instruct-q5_k_m-00002-of-00003.gguf")
hf_hub_download(repo, "qwen2.5-14b-instruct-q5_k_m-00003-of-00003.gguf")
llm   = QwenGGUF(file1)


executor = ThreadPoolExecutor(max_workers=4)

# ── app ───────────────────────────────────────────────────────────────────────
app = FastAPI()

# ── Pydantic schemas ──────────────────────────────────────────────────────────
class TextRequest(BaseModel):
    text: str = ""


# ── routes ────────────────────────────────────────────────────────────────────
@app.get("/")
def home():
    return {"status": "running", "model": MODEL_KEY}
    
llm_lock = asyncio.Lock()

@app.post("/transcribe")
async def transcribe(file: UploadFile = File(...)):

    if not file.filename:
        raise HTTPException(400, "empty_filename")

    suffix = Path(file.filename).suffix or ".mp3"

    with tempfile.NamedTemporaryFile(
        delete=False,
        suffix=suffix
    ) as tmp:
        tmp.write(await file.read())
        tmp_path = tmp.name


    async def event_generator():

        try:
            t0 = time.time()

            task = asyncio.get_running_loop().run_in_executor(
                executor,
                lambda: stt_model.transcribe(tmp_path)
            )


            while True:
                try:
                    result = await asyncio.wait_for(
                        asyncio.shield(task),
                        timeout=10
                    )

                    break

                except asyncio.TimeoutError:
                    # gửi heartbeat
                    yield b" "


            latency_ms = (time.time() - t0) * 1000


            data = {
                "text": result["text"],
                "segments": result["segments"],
                "language": result["language"],
                "confidence": result["confidence"],
                "no_speech_prob": result["no_speech_prob"],
                "language_probability": result["language_probability"],
                "latency_ms": round(latency_ms, 2)
            }


            yield json.dumps(data).encode("utf-8")


        except Exception as e:
            traceback.print_exc()
            yield json.dumps({
                "detail": str(e)
            }).encode("utf-8")


        finally:
            if os.path.exists(tmp_path):
                os.remove(tmp_path)


    return StreamingResponse(
        event_generator(),
        media_type="application/json"
    )



@app.post("/generate/summary")
async def generate_summary(body: TextRequest):

    if not body.text:
        raise HTTPException(400, "empty text")

    messages = [
        {
            "role": "system",
            "content": SUMMARY_SYSTEM
        },
        {
            "role": "user",
            "content": body.text
        }
    ]

    async def generator():

        t0 = time.time()

        try:
            loop = asyncio.get_running_loop()

            # lock + ping
            while True:
                try:
                    await asyncio.wait_for(
                        llm_lock.acquire(),
                        timeout=15
                    )
                    break

                except asyncio.TimeoutError:
                    yield b" "


            result = await loop.run_in_executor(
                executor,
                lambda: llm.generate(messages)
            )

            yield json.dumps({
                "summary": result,
                "latency_ms": round(
                    (time.time()-t0)*1000,
                    2
                )
            }).encode()


        finally:
            if llm_lock.locked():
                llm_lock.release()


    return StreamingResponse(
        generator(),
        media_type="application/json"
    )



@app.post("/generate/tasks")
async def generate_tasks(body: TextRequest):

    if not body.text:
        raise HTTPException(400, "empty text")


    messages = [
        {
            "role": "system",
            "content": TASK_SYSTEM
        },
        {
            "role": "user",
            "content": body.text
        }
    ]


    async def generator():

        t0 = time.time()

        try:
            loop = asyncio.get_running_loop()

            while True:
                try:
                    await asyncio.wait_for(
                        llm_lock.acquire(),
                        timeout=15
                    )
                    break

                except asyncio.TimeoutError:
                    yield b" "


            result = await loop.run_in_executor(
                executor,
                lambda: llm.generate(
                    messages,
                    temperature=0.0,
                    max_tokens=2048
                )
            )


            yield json.dumps({
                "tasks_raw": result,
                "latency_ms": round(
                    (time.time()-t0)*1000,
                    2
                )
            }).encode()


        finally:
            if llm_lock.locked():
                llm_lock.release()



    return StreamingResponse(
        generator(),
        media_type="application/json"
    )



@app.post("/diarize")
async def diarize(file: UploadFile = File(...)):
    if not file.filename:
        raise HTTPException(400, "empty_filename")
    try:
        content = await file.read()
        async def event_generator():
            t0 = time.time()
            
            # Khởi chạy hàm phân tách dưới background
            task = asyncio.get_running_loop().run_in_executor(
                executor,
                lambda: diarization_model.diarize(content)
            )
            # Vòng lặp chờ kết quả và nhả heartbeat (khoảng trắng)
            while True:
                try:
                    result = await asyncio.wait_for(
                        asyncio.shield(task),
                        timeout=10  # Cứ 10s ping 1 lần
                    )
                    break  # Xong rồi thì thoát vòng lặp
                except asyncio.TimeoutError:
                    yield b" "  # Nhả 1 byte trắng để giữ kết nối Cloudflare
            latency_ms = (time.time() - t0) * 1000
            # Đóng gói kết quả
            data = {
                "segments": result["segments"],
                "latency_ms": round(latency_ms, 2)
            }
            # Trả về chuỗi JSON thật ở cuối stream
            yield json.dumps(data).encode("utf-8")
        return StreamingResponse(
            event_generator(),
            media_type="application/json"
        )
    except Exception as e:
        import traceback
        traceback.print_exc()
        raise HTTPException(500, str(e))


@app.post("/generate/check-conflicts")
async def check_conflicts(body: TextRequest):
    if not body.text:
        raise HTTPException(400, "empty text")
    messages = [
        {
            "role": "system",
            "content": CHECK_CONFLICTS_SYSTEM
        },
        {
            "role": "user",
            "content": body.text
        }
    ]
    async def generator():
        t0 = time.time()
        try:
            loop = asyncio.get_running_loop()
            # Chờ lấy lock, gửi ping nếu vượt quá 15s để tránh Cloudflare timeout
            while True:
                try:
                    await asyncio.wait_for(
                        llm_lock.acquire(),
                        timeout=15
                    )
                    break
                except asyncio.TimeoutError:
                    yield b" "
            result = await loop.run_in_executor(
                executor,
                lambda: llm.generate(
                    messages,
                    temperature=0.0,  # Temperature = 0 để bắt LLM trả kết quả JSON cứng ngắc và chính xác nhất
                    max_tokens=1024
                )
            )
            # Trả về nguyên khối JSON từ LLM
            yield json.dumps({
                "conflicts": result,
                "latency_ms": round((time.time()-t0)*1000, 2)
            }).encode()
        finally:
            if llm_lock.locked():
                llm_lock.release()
    return StreamingResponse(
        generator(),
        media_type="application/json"
    )
```
